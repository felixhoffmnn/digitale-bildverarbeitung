## Importiere notwendige Bibliotheken

```python
import numpy as np
import pandas as pd
import cv2
from matplotlib import pyplot as plt
import scipy
```

## Lade Bild und zeige es in eigenem Fenster an

```python
# lade Bild und zeige es in eigenem Fenster an
img=cv2.imread('./img/lena_std.tif', cv2.IMREAD_COLOR)
cv2.imshow('Testbild Lena, www.lenna.org', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Lade Bild und zeige es "Inline" an

```python
# zeige Bild innerhalb des Notebooks an
%matplotlib inline
img = cv2.imread('./img/lena_std.tif', cv2.IMREAD_COLOR)
plt.figure()
plt.imshow(img)
# konvertiere Bildfarben
plt.figure()
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```
