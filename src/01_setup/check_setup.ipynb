## Importiere notwendige Bibliotheken

```python
from pathlib import Path

import cv2
import numpy as np
import pandas as pd
import scipy
from matplotlib import pyplot as plt
```

## Konfigurieren von Variablen

```python
img_dir = Path("../../data/images")
```

## Lade Bild und zeige es in eigenem Fenster an

```python
# lade Bild und zeige es in eigenem Fenster an
img = cv2.imread(str(img_dir / "lena_std.tif"), cv2.IMREAD_COLOR)
cv2.imshow("Testbild Lena, www.lenna.org", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Lade Bild und zeige es "Inline" an

```python
# zeige Bild innerhalb des Notebooks an
%matplotlib inline
img = cv2.imread(str(img_dir / "lena_std.tif"), cv2.IMREAD_COLOR)
plt.figure()
plt.imshow(img)
# konvertiere Bildfarben
plt.figure()
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

## Bildkoordinatensystem

```python
# Zeige die ersten 25 Spalten des Bildes an
plt.figure()
plt.imshow(img_rgb[:,0:24])
plt.show()
```

```python
# Anmerkung: plt.imshow(img[...], cmap='gray', vmin=0, vmax=255) ermglicht die Darstellung als Grauwert-Bild
# und verhindert durch vmin und vmax die automatische Skalierung der Pixelwerte
plt.figure()
imGray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
plt.imshow(imGray, cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
# Anmerkung: plt.imshow(img[...], cmap='gray', vmin=0, vmax=255) ermglicht die Darstellung als Grauwert-Bild
# und verhindert durch vmin und vmax die automatische Skalierung der Pixelwerte
plt.figure(figsize=(20, 10))
plt.subplot(1, 3, 1)
plt.imshow(img_rgb[:, :, 0], cmap="gray", vmin=0, vmax=255)
plt.subplot(1, 3, 2)
plt.imshow(img_rgb[:, :, 1], cmap="gray", vmin=0, vmax=255)
plt.subplot(1, 3, 3)
plt.imshow(img_rgb[:, :, 2], cmap="gray", vmin=0, vmax=255)
plt.show()
```
